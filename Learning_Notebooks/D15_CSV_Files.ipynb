### 3. Index_col Parameter

when you want to set a column in CSV as row index you set ```index_col``` parameter to the column name as string

````python
pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv", index_col = 'Country')
````

````python
pd.read_csv('datasets/cardio_train.csv',sep = ';',index_col='id')
````

### 4. Header parameter

When your column names are being treated as 1st row, you can set ```header``` paramerter to 1 (for first row)

if you set header to some row 'n', the data frame starts from row 'n+1'

`headers` and `header` are different Parameters

````python
pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv",header=0)
````

````python
pd.read_csv('datasets/cardio_train.csv',sep = ';', header=None)
````

### 5. ```usecols``` parameter

to import only specfic columns from csv file.

input is list of column names as strings.

````python
pd.read_csv("https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv", usecols=['Region'])
````

````python
pd.read_csv("datasets/cardio_train.csv",sep = ';', usecols=['age', 'gender', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio'])
````

### 7. Skiprows/nrows
skiprows:
skip specified row numbers in list, or using function

nrows:
import only first n rows

Can be used when lot of data is there and ur ram is not sufficient to load it.

````python
pd.read_csv("datasets/cardio_train.csv",sep = ';',skiprows=4)
# or
````

````python
pd.read_csv("datasets/cardio_train.csv", sep=';',skiprows = [2,4,6,8,10],nrows = 15)
# or
````

````python
pd.read_csv("datasets/cardio_train.csv", sep=';',skiprows = lambda x:x%2==1)
````

### 8. Encoding parameter

sometimes dataset is encoded in different encoding format. default and most used is UTF-8. sometimes it is latin-1, may be diffferent for hindi, emojis dataset, etc.

We can change encoding by using `encoding` parameter like `encoding = 'latin-1'`

### 9. Skip bad lines
`error_bad_lines = True`

to remove bad lines. like for csv_read suppose we have 8 cs values in each line but some line has 9 or 10 values randomly, it gives parser error. to avoid this error we use `error_bad_lines`

### 10. dtypes parameter

`dtype` parameter is used to change data type of specific column.
input is Dict with key value pairs of column name and new data type respectively.

````python
pd.read_csv("datasets/cardio_train.csv",sep = ';',dtype = {'alco': bool, 'height':float})

# pd.read_csv("datasets/cardio_train.csv",sep = ';',dtype = {'alco': 'bool', 'height':'float'}).info()
````

### 11. Handeling Dates

Dates in pandas normally are fetched in string format from csv, to convest it to DateTime64 format u can use this parameter as follows:
###### can use index or name of column too
`parse_dates = [...]` list of columns to be converted to dates

`parse_dates = [[...]]` list of list of columns to merge multiple columsn like date,month and year in single column

`parse_dates = {'foo':[...]}` can also give name to combined column like this. name = foo
